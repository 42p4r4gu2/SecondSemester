To calculate the surface area of the Earth, we can model the Earth as a sphere
or as oblate spheroid to approximate its potato-like shape, as shown in Figure
1.
The formula for the surface area of an oblate spheroid is given by: \
$S = 2\pi a^2 ( 1+\frac{1-e^2}{e}$tanh<sup>-1</sup>$(e))$

```python
import numpy as np
x = 2 * pi * (a ** 2) * (1 + ((1 - 3 ** 2) / e) * np.arctanh(e))
print(x)
```
