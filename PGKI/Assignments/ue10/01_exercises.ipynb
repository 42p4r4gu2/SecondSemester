### Exercise: Plotting Functions

Plot the functions $f_i(x^i)$ for $i=1,2,3$ on the interval $[-1.5, 1.5]$. Follow these steps:
1. Plot each function using a different line style.

2. Set the line width to 2 for each function.

3. Display a legend in the lower right corner of the plot.

4. Add a descriptive title to the plot.

5. Label the x- and y-axes.

6. Add a grid to the plot.

```python
# code
x = np.linspace(-1.5, 1.5, 100)
title = ["x", "x squared", "x cubic"]
ls = [":", "--", "-"]
y_labels = ["x", "x^2", "x^3"]
f = [x, x**2, x**3]
figs, axes = plt.subplots(1,3, figsize=(13,2))
for i, ax in enumerate(axes.flatten()):
    ax.plot(x, f[i], lw=2, label=title[i], linestyle=ls[i])
    ax.set_title(title[i])
    ax.legend(loc=4)
    ax.set(title=title[i], ylabel=y_labels[i], xlabel="x")
    ax.grid()
```

### Exercise

Plot the functions $x$, $x^2$, ..., $x^6$ on the interval $[-2,+2]$ in a 2 x 3 grid. Use a figure size of 14 x 12.

Follow these steps:
1. Plot each function in a separate subplot of the grid.

2. Set the title of each subplot to 'f(x) = x^i', where 'i' is the corresponding power of the function.

3. Label the x- and y-axes of each subplot.

4. Set the line width to 3 for each function.


```python
# code
figs, axes = plt.subplots(2, 3, figsize=(14,12))
x = np.linspace(-2,2,100)
for i, ax in enumerate(axes.flatten()):
    ax.plot(x, x**(i+1), lw = 3)
    ax.set(xlabel="x", ylabel="f(x)")
    ax.set(title=f"f(x):x^{i+1}")
```
