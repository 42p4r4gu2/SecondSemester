## 2. Simple Plots


One way to use matplotlib is through its 'functional interface', provided by the `matplotlib.pyplot` module. This interface allows you to create plots quickly and easily by managing the current figure and axes objects automatically. With just a few lines of code, you can add plot elements such as lines, markers, and text to an Axes, and matplotlib takes care of the rest. This makes it easy to generate plots interactively and explore your data visually.

````python
import numpy as np
import matplotlib.pyplot as plt
````

### 2.1. Line and Scatter Plots

Content:
+ line plots
+ scatter plots
+ saving plots

#### Line plot: sine and cosine

````python
x = np.linspace(-np.pi, np.pi, 100)

plt.plot(x, np.cos(x))
plt.plot(x, np.sin(x))
plt.show()
````

+ `plt.plot` 

    + plots y versus x as a line

    + x and y are two iterable objects of the same length

    + creates a Matplotlib object (here, a Line2D object)

+ `plt.show` 

    + displays open figure on the screen

    + can be omitted in Jupyter notebook

#### Scatter plot

````python
# matrix of 100 random 2D points 
X = np.random.uniform(-1, 1, (100, 2))
plt.scatter(X[:,0], X[:,1])
plt.show()
````

#### Saving a plot
Save plot as an image 
```python
    plt.savefig('plot.png')        # save as .png-image
    plt.savefig('plot.pdf')        # save as .pdf-file
    plt.savefig('plot.eps')        # save as .eps-file
```
<br>

`plt.savefig(filename)`

+ call at the end but before `plt.show`

+ image format is deduced from file extension

````python
x = np.random.uniform(-1, 1, (100, 2))
plt.scatter(x[:,0], x[:,1])
plt.savefig('plot.png')
plt.show()
````

### 2.2. Labels and Legends

The next script plots the sine and cosine with legends:

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine with label
plt.plot(x, np.cos(x), label="cos")

# plot sine with label
plt.plot(x, np.sin(x), label="sin")

# places legend 
plt.legend()

# show result on screen
plt.show()
````

**Description:**

+ each line can be labelled by passing a string object to its `label` argument

+ call `plt.legend` to add legend to the plot

+ note: without `plt.legend` no legend appears on the plot

<br>

**Location of legend**

+ can be customized by setting the `loc` argument of the `plt.legend` method

+ argument can be either a string or an integer as shown in the table below

<br/>

|'best' | 0
|'upper right'| 1 
|'upper left'| 2 
|'lower left'| 3 
|'lower right'| 4 
|'center right'| 5 
|'center left'| 6 
|'center right'| 7

#### Setting the location of the legend

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine with label
plt.plot(x, np.cos(x), label="cos")

# plot sine with label
plt.plot(x, np.sin(x), label="sin")

# sets location by passing a string to argument loc
plt.legend(loc='lower right')

# sets location by passing an integer to argument loc
#plt.legend(loc=2)

plt.show()
````

### 2.3. Title and Axis Labels

The next script plots a title above the plot and labels to annotate the x- and y-axes. We use the following functions:

+ `plt.title`

+ `plt.xlabel`

+ `plt.ylabel`


````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine with label
plt.plot(x, np.cos(x), label="cos")

# plot sine with label
plt.plot(x, np.sin(x), label="sin")

# set title
plt.title("sine and cosine")

# set label for x-axis
plt.xlabel("x")

# set label for y-axis
plt.ylabel("f(x)")

plt.legend()
plt.show()
````

### 2.4. Line Styles and Widths

By default, `plot` produces solid lines of weight 1.5 pt. Style and width can be both customized.

#### Customizing line style

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine as dashed line
plt.plot(x, np.cos(x), linestyle='--')

# plot sine with as dotted line
plt.plot(x, np.sin(x), linestyle='dotted')

plt.show()
````

**Linestyles**: Linestyle specifiers to set the `linestyle` argument.

| symbol | string   | description
|:------:|:---------|:-----------
| '-'    | 'solid'  | solid line
| '--'   | 'dashed' | dashed line
| '-.'   | 'dashdot'| dash-dotted line
| ':'    | 'dotted' | dotted line
| ' '    | 'None‘   | draw nothing

Linestyles can be specified in different ways:

+ Variant 1: `plot(x, y)` plots a solid line

+ Variant 2: `plot(x, y, ':')` plots a dotted line

+ Variant 3: `plot(x, y, linestyle=':')` plots a dotted line

+ Variant 4: `plot(x, y, linestyle='dotted')` plots a dotted line

Variant 1 is the default style and Variant 2 is an abbreviation. 

#### Customizing line width

The next script customizes the width of a line.

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine as dashed line with linewidth 3.5
plt.plot(x, np.cos(x), linestyle='--', linewidth=3.5)

# plot sine with as dotted line with linewidth 2
plt.plot(x, np.sin(x), linestyle=':', linewidth=10)

plt.show()
````

**Description:**
Specify the thickness (in points) of a line by setting the `linewidth` argument to a non-negative float.

### 2.5. Markers

By default, `plot` produces lines without markers at the defining points. The next script will plot lines with markers:

````python
# Note: only 20 points are generated for the sake of better illustration

# create data
x = np.linspace(-np.pi, np.pi, 20)

# plot cosine as dashed line with marker
plt.plot(x, np.cos(x), '--o')

# plot sine with as dotted line with marker
plt.plot(x, np.sin(x), linestyle=':', marker='*')

plt.show()
````

The [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) documentation lists all markers and describes how to handle markers.

### 2.6. Color

Next, we will set the color of the lines and markers:

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine as red dashed line 
plt.plot(x, np.cos(x), linestyle='--', color='red')

# plot sine with as blue dotted line
plt.plot(x, np.sin(x), linestyle=':', color='tab:blue')

plt.show()
````

**Description:**
+ set the `color` argument 
+ several formats for specifying the colore are supported

**Note:** marker fill and edge color can also be set.

<br/>

**Color codes**
|code | string  | tableau colors
|-----|---------|---------------
| b   | blue    | tab:blue
| g   | green   | tab:orange
| r   | red     | tab:green
| c   | cyan    | tab:red
| m   | magenta | tab:purple
| y   | yellow  | tab:brown 
| k   | black   | tab:pink
| w   | white   | tab:gray
|     |         | tab:olive
|     |         | tab:cyan

<br>

#### Formats

Matplotlib supports several formats for specifying the color:

+ **Format 1:** `plot(x, y, color='r')`

    + basic color code 
    
    + see table

+ **Format 2:** `plot(x, y, color='red')`

    + basic color string 
    
    + see table

+ **Format 3:** `plot(x, y, color='tab:red')`

    + tableau color string

    + default color sequence

+ **Format 4:** `plot(x, y, color='value')`

    + `value` is a float between 0. and 1. passed as a string

    + example: `color='0.7'`

    + values represent shades of gray

    + black = 0.

    + white = 1.

+ **Format 5:** `plot(x, y, color=HEX)`

    + HEX is a string of the form '#RGB'

    + R,G,B are hex-numbers in the range 00-ff

    + components: R = red, G = green, B = blue

    + example: `color='#ff00ff'` is magenta

+ **Format 6:** `plot(x, y, color=rgb)`

    + rgb is a tuple (r, g, b) of floats in the range 0 - 1

    + example: `color=(0.5, 0., 0.)` is dark red

### 2.7. Axis Limits

Methods `plt.xlim` and `plt.ylim` 

+ set the x- and y-limits of the plot, respectively

+ must be called *after* any `plt.plot` statements

+ must be called before showing or saving the figure

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# plot cosine and sine 
plt.plot(x, cos_x, x, sin_x)

plt.xlim(1.25*np.min(x), 1.25*np.max(x))
plt.ylim(1.25*np.min(cos_x), 1.25*np.max(cos_x))

plt.show()
````

### 2.8. Axes Ticks

Methods `plt.xticks` and `plt.yticks` set the tick locations and labels of the x- and y-axis, respectively. 

The next script will set the ticks to the characteristic points of sine and cosine:

+ xticks: $-\pi, -\pi/2, 0, \pi/2, \pi$ 

+ y-ticks: $-1, 0, 1$

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine and sine 
plt.plot(x, np.cos(x), x, np.sin(x))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])

plt.show()
````

**Remark:** Observe that the x-labels are the numbers $-3.142, -1.571, \ldots$. It would be more readable to label the x-ticks with the symbols $-\pi, -\pi/2, \ldots$. The appendix shows how to render text in a plot.

### 2.9. Abbreviations

**a)** Abbreviations for plot line properties:

+ `c` for color

+ `ls` for linestyle

+ `lw` for linewidth

**Example:**
```python
    plt.plot(x, y, c='g', ls='--', lw=5)    # green, dashed, thick line
```

<br/>

**b)** Abbreviations by single strings that specify 

+ color

+ line style

+ marker style

**Example:**
```python
    plt.plot(x, y, 'r:o')    # red, dotted line with circle markers
    plt.plot(x, y, '--b')    # blue, dashed line without markers
```

<br>

Order and number of specifiers in the string are arbitrary:

````python
# create data
x = np.linspace(-np.pi, np.pi, 20)
cos_x, sin_x = np.cos(x), np.sin(x)

# plot cosine with red markers
plt.plot(x, cos_x, 'ro')

# plot sine with as blue dashed line with markers
plt.plot(x, sin_x, '*--b')

plt.show()
````

**c)** Multiple lines can be plotted using a sequence of x, y format arguments:

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)

# plot cosine and sine 
plt.plot(x, np.cos(x), 'r--', x, np.sin(x), 'b-.')
plt.show()
````

### Appendix: LaTeX in `pyplot`

You can use LaTeX to render text in matplotlib plots within a Jupyter notebook. To do this, you need to have a working LaTeX installation on your system, as well as the other dependencies described in the matplotlib documentation [[>](https://matplotlib.org/stable/tutorials/text/usetex.html)].

**Usage:** Enable option

```python
    plt.rc('text', usetex=True)
```
for using LaTeX in pyplot plots. The method `pyplot.rc` is used to set runtime configuration parameters. Use raw strings `r'abc'` to prevent Python from escaping characters followed by LaTeX's backslahes.

**Example:**

````python
# create data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# plot cosine and sine 
plt.plot(x, cos_x, x, sin_x)

# store ticks and labels in lists using LaTeX
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
yticks = [-1, 0, +1]

# LaTeX: store labels in lists 
xticklabels = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$']
yticklabels = ['$-1$', '$0$', '$+1$']
               
# set ticks and labels
plt.xticks(xticks, xticklabels)
plt.yticks(yticks, yticklabels)

# LaTeX: title and axis labels
plt.title(r'$\cos(x)$ and $\sin(x)$' )
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')

plt.show()
````
